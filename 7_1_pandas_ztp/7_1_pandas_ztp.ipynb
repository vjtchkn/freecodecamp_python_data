### Analyzing data from data frames

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
print("Total cases:", total_cases)
print("Total deaths:", total_deaths)
```

```python
death_rate = covid_df.new_deaths.sum() / covid_df.new_cases.sum()
print("Reported death rate:", round(death_rate * 100, 2), "%")
```

```python
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()
total_tests
```

```python
positive_rate = total_cases / total_tests
print(f"Positive rate: {positive_rate * 100:.2f} %")
```

### Grouping and aggregation

```python
covid_df
```

```python
monthly_groups = covid_df.groupby("month")
monthly_groups[["new_cases", "new_deaths", "new_tests"]].sum()
```

```python
covid_weekday_mean_df = covid_df.groupby("weekday")[["new_cases", "new_deaths", "new_tests"]].mean()
covid_weekday_mean_df
```

```python
covid_df["total_cases"] = covid_df.new_cases.cumsum()
covid_df["total_deaths"] = covid_df.new_deaths.cumsum()
covid_df["total_tests"] = covid_df.new_tests.cumsum() + initial_tests
covid_df
```

### Merging data from multiple sources

```python
urlretrieve('https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv', 
            'locations.csv')
```

```python
locations_df = pd.read_csv("locations.csv")
locations_df
```

```python
locations_df[locations_df.location == "Italy"]
```

```python
covid_df["location"] = "Italy"
covid_df
```

```python
merged_df = covid_df.merge(locations_df, on="location")
merged_df
```

```python
merged_df["cases_per_million"] = merged_df.total_cases / merged_df.population * 1000000
merged_df["deaths_per_million"] = merged_df.total_deaths / merged_df.population * 1000000
merged_df["tests_per_million"] = merged_df.total_tests / merged_df.population * 1000000
merged_df
```

### Writing data to file

```python
results_df = merged_df[["date",
                        "new_cases",
                        "total_cases",
                        "new_deaths",
                        "total_deaths",
                        "new_tests",
                        "total_tests",
                        "cases_per_million",
                        "deaths_per_million",
                        "tests_per_million"]]
results_df
```

```python
results_df.to_csv("results.csv", index=False)
```
